```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\Raman\Downloads\householdtask3.csv")
data
```

```python
data.head(10)  # To show the top 10 records of a dataset
```

```python
#Scatter plot with year against own 

plt.scatter(data['year'],data['own'])

#Adding title to the plot
plt.title("Sactter Plot")

# setting the X and Y labels
plt.xlabel('year')
plt.ylabel('own')

# showing the Chart
plt.show()
```

```python
# Line Chart Year against Own 

plt.plot(data['year'])
plt.plot(data['own'])

plt.title("Line chart")

plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
# Bar chart or Bar plot 

plt.bar(data['year'],data['own'])

plt.title("Bar chart")

plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
# Histogram 

plt.hist(data['income'])

plt.title("Histogram")

plt.show
```
